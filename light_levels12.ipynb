```python
import pennylane as qml
from qco_new import *
```

```python
[ds] = qml.data.load("ketgpt")
```

```python
# example
seed = 0
@qml.qnode(qml.device('default.qubit'))
def circuit():
    for op in ds.circuits[seed]:
        if op.name == 'QubitUnitary':
            continue
        else:
            qml.apply(op)
    return qml.state()

show_circuit(circuit)
```

```python
info1 = optimize12q(qnode=circuit)
show_info(info1)
```

```python
info2 = optimize1q(circuit_info=info1)
show_info(info2)
```

```python
info3 = optimize1q(circuit_info=replace_gate(info2))
show_info(info3)
```
